```python
# Create CSV file
with open('tickets_generated_2.csv', mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(["Ticket_ID", "Date", "Machine", "Description"])  # Header
    
    start_date = datetime(2023, 1, 1)
    tickets_per_day = 3 
    total_tickets = 1000
    
    ticket_id = 1
    for i in range(total_tickets):
        # 3 tickets for one day
        current_date = start_date + timedelta(days=(i // tickets_per_day))
        row = [
            ticket_id,  
            current_date.date(), 
            random.choice(machines), 
            random.choice(descriptions) 
        ]
        writer.writerow(row)
        ticket_id += 1
```
